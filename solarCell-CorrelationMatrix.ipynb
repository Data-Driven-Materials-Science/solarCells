```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col_list = ['temperature', 'speed', 'distance', 'pressure', 'efficiency']
dataset = pd.read_csv('deviceSummary.csv', usecols=col_list)
print(dataset)
```

```python
corrMatrix = dataset.corr()
print(corrMatrix)
import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(corrMatrix, annot=True, low=-1)
plt.title("Correlation Matrix of Independent and \nDependent (efficiency) Varaibles for \n Solar Cell Production")
plt.show()
```
