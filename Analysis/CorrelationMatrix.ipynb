# Correlation matrix
This Notebook is designed to take a given dataset and make a coorelation matrix for it

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col_list = ['Temperature', 'Distance', 'Speed', 'Pressure', 'Efficiency']
#dataset = pd.read_csv('../DataSets/standardDeviceSummary.csv', usecols=col_list)
#dataset = pd.read_csv('../DataSets/newDataCollection.csv', usecols=col_list)
col_list = ['Temperature', 'Distance', 'Speed', 'Pressure', 'Efficiency']
dataset = pd.read_csv('../DataSets/optimizedSyntheticDataset.csv', usecols=col_list)
print(dataset)
```

```python
corrMatrix = dataset.corr()
print(corrMatrix)
import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(corrMatrix, annot=True, vmin=-1)
plt.title("Correlation Matrix of Independent and \nDependent (efficiency) Variables for \n Solar Cell Production")
plt.show()
```
