```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
col_list = ['temperature', 'speed', 'distance', 'pressure', 'efficiency']
#col_list = ['temperature', 'speed', 'distance', 'pressure']
dataset = pd.read_csv('../deviceSummary.csv', usecols=col_list)
normalizedDataset = normalize(dataset)
print(normalizedDataset)
```

```python
covMatrix = pd.DataFrame.cov(normalizedDataset)
print(covMatrix)
sn.heatmap(covMatrix, annot=True)
plt.title("Covariance Matrix of Independent and \nDependent (efficiency) Variables for \n Solar Cell Production. *Data Normalized*")
plt.show()
```

```python
covMatrix = pd.DataFrame.cov(dataset)
print(covMatrix)
sn.heatmap(covMatrix, annot=True, vmax=50)
plt.title("Covariance Matrix of Independent and \nDependent (efficiency) Variables for \n Solar Cell Production. *Color Max set to 50*")
plt.show()
```
