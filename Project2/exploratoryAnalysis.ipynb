```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from an Excel file
df = pd.read_excel('Solar_cell_Devices.xlsx')
df = df.drop(columns=['Temperature'])
# raw_df = raw_df.drop(columns=['Training / Validation'])
# raw_df = raw_df[raw_df['Width'] != 'TBD']
# raw_df = raw_df[raw_df['Width'] != -1]

# # Drop three columns (e.g., 'Column1', 'Column2', 'Column3')
# columns_to_drop = ['Substrate Type', 'Wire']
# df = raw_df.drop(columns=columns_to_drop)

print(df.shape)
print(df)
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"fontsize": 35})
# Customize font size for x-axis and y-axis labels
heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=20)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=20)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
